```python
## ReadFile
date_cols = ["DATEANDTIME"]
df =  pd.read_csv('HISEVENT_2019_12_17_00_00_00', header=0, skiprows=[1], sep=';' , on_bad_lines='skip' ,parse_dates=date_cols)
## print(df)
```

```python
## Describe
##df['DATEANDTIME'].astype('datetime64')
df.dtypes
df.info()
```

```python
## Drop Cols
df = df.drop(columns=[ '#ALARMID', 'ENVIRONMENT', 'VALUE', 'ACKREQUIRED', 'SEVERITY', 'EQUIPMENTCLASS', 'FUNCTIONALCAT', 'GEOGRAPHICALCAT', 'GROUP1', 'GROUP2', 'FORMAT'	])
df.head()
```

```python
## Summary Count
df.groupby("ASSETNAME")["ASSETNAME"].count()
```

```python

def filter_func(x):
    return x['EQUIPMENTNAME'].str.contains('SMC__0001', na=False) 

df.groupby(['EQUIPMENTNAME']).filter(filter_func)
## df.groupby("EQUIPMENTNAME").filter(lambda x: x['EQUIPMENTNAME'].str.contains('SMC__0001', na=False) ).count()
```
